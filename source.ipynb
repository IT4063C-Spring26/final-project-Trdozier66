Problem:
Many college students work part-time or full-time while enrolled in school, but it is not always clear how employment affects academic performance and student success. Why this topic is important:
Understanding the relationship between student employment and academic performance can help students make informed decisions about work hours and course loads. It can also help colleges and universities design better support systems, advising strategies, and financial aid policies to improve student success and retention.

Data Sources
1. National Postsecondary Student Aid Study (NPSAS)

Type: File (CSV)

Source: U.S. Department of Education

Description: This dataset provides student-level information including employment status, hours worked per week, GPA, enrollment intensity, and demographic characteristics.

Use in Project: This dataset will be used to analyze how student employment and work hours relate directly to academic performance.

2. Integrated Postsecondary Education Data System (IPEDS)

Type: Database / File (CSV)

Source: National Center for Education Statistics (NCES)

Description: IPEDS contains institution-level data such as retention rates, graduation rates, enrollment size, and institutional characteristics.

Use in Project: This dataset will provide institutional context to compare academic outcomes across different types of colleges and universities.

3. College Scorecard API

Type: API

Source: U.S. Department of Education

Description: The College Scorecard API provides school-level metrics including completion rates, student demographics, and post-graduation outcomes.

Use in Project: This dataset will supplement institutional performance data and allow comparison across schools.

How the Datasets Will Be Related

The datasets will be related using institutional identifiers such as school ID codes and institution names. Student employment data from NPSAS will be analyzed alongside institutional-level outcomes from IPEDS and College Scorecard to examine how employment patterns correlate with academic performance and retention across different colleges.
